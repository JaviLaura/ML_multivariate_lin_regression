Import csv

```python
df = pd.read_csv('hiring.csv')
display(df)
```

Fill null values in 'experience' column with zero assuming that no value entered means zero.

```python
df.experience = df.experience.fillna(0)
display(df)
```

Convert 'experience' column values to numbers.

```python
df['experience'] = df['experience'].apply(lambda x: w2n.word_to_num(str(x)))
display(df)
```

Fill null values in 'test_score' column with median.

```python
import math
median_test_score = math.floor(df.test_score.median())
median_test_score
```

```python
df.test_score = df.test_score.fillna(median_test_score)
display(df)
```

Formula:

y = m1x1 + m2x2 + m3x3 + b

m1, m2, and m3 are 'coefficents

b is 'intercept'

'Dependent variable' is salary.

'Features' or 'independent variables' are experience, test_score, and interview_score.

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score', 'interview_score']],df.salary)
```

Look at coefficients

```python
reg.coef_
```

Look at intercept

```python
reg.intercept_
```

Predict salary for someone with 2 years experience, 9 for test score, and 6 for interview score

```python
reg.predict([[2, 9, 6]])
```

Formula to double check above prediction:

```python
# price = m1 * experience + m2 * test_score + m3 * interview_score + b
2812.95487627 * 2 + 1845.70596798 * 9 + 2205.24017467 * 6 + 17737.263464337688
```
